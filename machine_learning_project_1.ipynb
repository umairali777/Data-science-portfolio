```python
NAs = pd.concat([train.isnull().sum()], axis=1, keys=["train"])
NAs[NAs.sum(axis=1) > 0]
```

i'm going to fill missing vlalues of "AGE" wth mean values of age in train datset and Cabin and embarked values with mode

```python
train["Age"] = train["Age"].fillna(train["Age"].mean())
```

```python
train["Cabin"] = train["Cabin"].fillna(train["Cabin"].mode()[0])
```

```python
train["Emarked"] = train["Embarked"].fillna(train["Embarked"].mode()[0])
```

```python
train["Pclass"] = train["Pclass"].apply(str)
```

```python
for col in train.dtypes[train.dtypes == "object"].index:
  for_dummy = train.pop(col)
  train = pd.concat([train, pd.get_dummies(for_dummy, prefix=col)], axis=1)
train.head()
```

in above cell we make dummies of those column which have categorical variables 

```python
labels = train.pop("Survived")
```

WE JUST REMOVED sURVIVED LABEL FROM TRAIN DATSET

```python
from sklearn.model_selection import train_test_split
x = [['passengerId', 'Age', 'Name', 'Sex','Sibsp', 'Pclass', 'Cabin', 'Parch', 'Fare', 'Ticket', 'Embarked' ]]
y= [['Survived']]
x_train, x_test, y_train, y_test = train_test_split(train, labels, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
from sklearn.metrics import roc_curve, auc
false_positive_rate, true_positive_rate, thersholds = roc_curve(y_test, y_pred)
roc_auc = auc(false_positive_rate, true_positive_rate)
roc_auc
```
